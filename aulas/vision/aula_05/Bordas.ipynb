```python
import cv2
import numpy as np

import pylab as plt
%matplotlib inline
```

```python
image = cv2.imread('../db_images/png/captcha.png', 0)
```

```python
plt.title("Original")
plt.xticks([]), plt.yticks([])
plt.imshow(image, cmap="gray")

plt.tight_layout()
```

```python
sobel_x = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)

plt.title("Sobel X")
plt.xticks([]), plt.yticks([])
plt.imshow(sobel_x, cmap="gray")

plt.tight_layout()
```

```python
sobel_y = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
plt.title("Sobel Y")
plt.xticks([]), plt.yticks([])
plt.imshow(sobel_y, cmap="gray")

plt.tight_layout()
```

```python
sobel_or = cv2.bitwise_or(sobel_x, sobel_y)
plt.title("Sobel or")
plt.xticks([]), plt.yticks([])
plt.imshow(sobel_or, cmap="gray")

plt.tight_layout()
```

```python
canny = cv2.Canny(image, 100, 255)
```

```python
plt.title("Canny")
plt.xticks([]), plt.yticks([])
plt.imshow(canny, cmap="gray")

plt.tight_layout()
```
